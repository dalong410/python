파이썬 대표적인 시각화 도구 : Matplotlib
%matplotlib inline 명령은 그래프의 결과를 출력물에 포함시켜라는 의미.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure
plt.plot([1,2,3,4,5,6,7,8,9,8,7,6,5,4,3,2,1,0])
plt.show()
```

```python
import numpy as np
t = np.arange(0,12,0.01)
y = np.sin(t)
```

```python
plt.figure(figsize=(10,6))
plt.plot(t, y)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(t, y)
plt.grid()
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Example of sinwave')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(t, np.sin(t), label='sin')
plt.plot(t, np.cos(t), label='cos')
plt.grid()
plt.legend()
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Example of sinwave')
plt.show()
```

```python
t = [0,1,2,3,4,5,6]
y = [1,4,5,8,9,5,3]
plt.figure(figsize=(10,6))
plt.plot(t, y, color='pink')
plt.show()
```

```python
t = [0,1,2,3,4,5,6]
y = [1,4,5,8,9,5,3]
plt.figure(figsize=(10,6))
plt.plot(t, y, color='pink', linestyle='dashed')
plt.show()
```

```python
t = [0,1,2,3,4,5,6]
y = [1,4,5,8,9,5,3]
plt.figure(figsize=(10,6))
plt.plot(t, y, color='pink', linestyle='dashed', marker='o')
plt.show()
```

```python
t = [0,1,2,3,4,5,6]
y = [1,4,5,8,9,5,3]
plt.figure(figsize=(10,6))
plt.plot(t, y, color='red', linestyle='dashed', marker='o', \
         markerfacecolor='blue', markersize=9)
plt.show()
```
